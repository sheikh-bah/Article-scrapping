# opencodez

```python
from requests import get
```

```python
html= 'https://www.opencodez.com/page/0'
response=get(html)
print(response.text)
```

```python
from bs4 import BeautifulSoup
from pandas import DataFrame
import pandas
import bs4
```

```python
soup=BeautifulSoup(response.content, 'html.parser')
type(soup)
print(soup)
```

# use prettify to organize our html text

```python
print(soup.prettify()[:10000])
```

```python
html_title=soup.findAll("h2", {"class":"title"})
len(html_title)
```

```python
for title in range(22):
    print(html_title[title].a['title'])
```

```python
for link in range(15):
    print(html_title[link].a['href'])
```

# collect short description of post

```python
short_desc=soup.findAll('div',{"class":'post-content image-caption-format-1'})
len(short_desc)
for post in range(12):
    #print((short_desc[post]).p)
    print((short_desc[post]).p.text.strip())
```
